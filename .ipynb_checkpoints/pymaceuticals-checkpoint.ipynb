```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv("data/mouse_drug_data.csv")
ct_df = pd.read_csv("data/clinicaltrial_data.csv")

# Combine the data into a single dataset
combined_data = pd.merge(mouse_df, ct_df, on="Mouse ID")
combined_df = pd.DataFrame(combined_data)

# Display the data table for preview
combined_df.head(10)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped = combined_df.groupby(['Drug', 'Timepoint'], as_index=False)['Tumor Volume (mm3)'].mean()
means = combined_df['Tumor Volume (mm3)']

# Convert to DataFrame
grouped_df = pd.DataFrame(grouped)
timeline = grouped_df.iloc[0:10, 1]
tumor_mean = grouped_df.pivot(index='Timepoint', columns='Drug', values=['Tumor Volume (mm3)'])

grouped_df.head(10)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint 
tumor_sem = combined_df.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem()

# Convert to DataFrame 
tumor_sem_df = pd.DataFrame(tumor_sem).reset_index()
tumor_sem_df.rename(columns = {'Tumor Volume (mm3)': 'SEM'}, inplace = True)
tumor_sem_reformat = tumor_sem_df.pivot(index='Timepoint', columns='Drug', values='SEM')
grouped_df_sem = pd.merge(grouped_df, tumor_sem_df, on=["Timepoint", "Drug"])

# Preview DataFrame
tumor_sem_reformat.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_avg = grouped_df_sem.pivot(index='Timepoint', columns='Drug', values=['Tumor Volume (mm3)'])
tumor_err = grouped_df_sem.pivot(index='Timepoint', columns='Drug', values=['SEM'])

tumor_avg.head(15)
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
ms_grp = combined_df.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean()

# Convert to DataFrame
ms_df = pd.DataFrame(ms_grp).reset_index()

# Preview DataFrame
ms_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
ms_sem_grp = combined_df.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem()
ms_sem = combined_df['Metastatic Sites']

# Convert to DataFrame
ms_sem_df = pd.DataFrame(ms_sem_grp).reset_index()

# Preview DataFrame
ms_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
ms_mean = ms_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
ms_err = ms_sem_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')

# Preview that Reformatting worked
ms_mean.head()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_grp = combined_df.groupby(['Drug', 'Timepoint'], as_index=False)['Mouse ID'].count()

#Convert to Mouse Survival Rate
mouse_grp['Mouse ID'] = (mouse_grp['Mouse ID'] * 100)/25
mouse_grp.rename(columns={'Mouse ID': 'Mouse Survival Rate'}, inplace = True)

# Convert to DataFrame
mouse_grp_df = pd.DataFrame(mouse_grp)

# Preview DataFrame
mouse_grp_df.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_df = mouse_grp_df.pivot(index='Timepoint', columns='Drug', values='Mouse Survival Rate')

# Preview the Data Frame
mouse_df.head()
```

```python
# Generate the Plot (Accounting for percentages)
fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, figsize = [8, 14])
fig.suptitle("Tumor, Metastatic Responses, and Survival Rate", fontsize=16, fontweight="bold")
plt.xlabel("Time (Days)", fontsize=14, fontweight="bold")

ax1.errorbar(timeline, tumor_avg['Tumor Volume (mm3)']["Capomulin"], yerr=tumor_err['SEM']["Capomulin"], linewidth=1, linestyle="solid", marker="o", markersize=7.5)
ax1.errorbar(timeline, tumor_avg['Tumor Volume (mm3)']["Infubinol"], yerr=tumor_err['SEM']["Infubinol"], linewidth=1, linestyle="solid", marker="v", markersize=7.5)
ax1.errorbar(timeline, tumor_avg['Tumor Volume (mm3)']["Ketapril"], yerr=tumor_err['SEM']["Ketapril"], linewidth=1, linestyle="solid", marker="p", markersize=7.5)
ax1.errorbar(timeline, tumor_avg['Tumor Volume (mm3)']["Placebo"], yerr=tumor_err['SEM']["Placebo"], linewidth=1, linestyle="solid", marker="*", markersize=7.5)
ax1.set_xlim(min(timeline)-2, max(timeline)+2)
ax1.grid(color='cyan')
ax1.set_ylabel("Average Tumor Volume (mm3)", fontsize=12, fontweight="bold")
ax1.legend(loc="best", fontsize="small", fancybox=True)

ax2.errorbar(timeline, ms_mean["Capomulin"], yerr=ms_err["Capomulin"], linewidth=1, linestyle="solid", marker="o", markersize=7.5)
ax2.errorbar(timeline, ms_mean["Infubinol"], yerr=ms_err["Infubinol"], linewidth=1, linestyle="solid", marker="v", markersize=7.5)
ax2.errorbar(timeline, ms_mean["Ketapril"], yerr=ms_err["Ketapril"], linewidth=1, linestyle="solid", marker="p", markersize=7.5)
ax2.errorbar(timeline, ms_mean["Placebo"], yerr=ms_err["Placebo"], linewidth=1, linestyle="solid", marker="*", markersize=7.5)
ax2.grid(color='cyan')
ax2.set_ylabel("Average Metastatic Sites", fontsize=12, fontweight="bold")
ax2.legend(loc="best", fontsize="small", fancybox=True)

ax3.errorbar(timeline, mouse_df["Capomulin"], linewidth=1, linestyle="solid", marker="o", markersize=7.5)
ax3.errorbar(timeline, mouse_df["Infubinol"], linewidth=1, linestyle="solid", marker="v", markersize=7.5)
ax3.errorbar(timeline, mouse_df["Ketapril"], linewidth=1, linestyle="solid", marker="p", markersize=7.5)
ax3.errorbar(timeline, mouse_df["Placebo"], linewidth=1, linestyle="solid", marker="*", markersize=7.5)
ax3.grid(color='cyan')
ax3.set_ylabel("Survival Rate (%)", fontsize=12, fontweight="bold")
ax3.legend(loc="best", fontsize="small", fancybox=True)

fig.savefig("Fig1.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
total_pct_change = ((tumor_avg['Tumor Volume (mm3)'].iloc[-1] - tumor_avg['Tumor Volume (mm3)'].iloc[0]) / tumor_avg['Tumor Volume (mm3)'].iloc[0]) * 100
# Display the data to confirm
total_pct_change_df = pd.DataFrame(total_pct_change).reset_index()
total_pct_change_df.rename(columns = {0: 'Percent Change'}, inplace = True)
total_pct_change_df.head()
```

```python
# Store all Relevant Percent Changes into a Tuple
relevant_pct_change = total_pct_change_df.loc[(total_pct_change_df['Drug'] == 'Capomulin') | (total_pct_change_df['Drug'] == 'Infubinol') | (total_pct_change_df['Drug'] == 'Ketapril') | (total_pct_change_df['Drug'] == 'Placebo')] 

# Splice the data between passing and failing drugs
passing_df = relevant_pct_change.loc[relevant_pct_change['Percent Change'] > 0].set_index('Drug')
failing_df = relevant_pct_change.loc[relevant_pct_change['Percent Change'] < 0].set_index('Drug')

# from matplotlib.ticker import MultipleLocator, FormatStrFormatter
fig = plt.figure()
ax = fig.add_subplot(111)

bars = plt.bar(failing_df.index, failing_df['Percent Change'], width=1, color='g', alpha=0.5, align='center')
bars2 = plt.bar(passing_df.index, passing_df['Percent Change'], width=1, color='r', alpha=0.5, align='center')
plt.title('Tumor Change Over 45 Day Treatment')
plt.ylabel('% Tumor Volume Change')
plt.xlabel('Treatment Type')
ticks = np.asarray([i + 0.5 for i in range(len(relevant_pct_change.index))])
labels = np.asarray([i for i in range(len(relevant_pct_change.index))])

plt.xticks(ticks)
ax.set_xticks(labels, minor=True)
ax.set_xticklabels(relevant_pct_change["Drug"])
plt.grid(color='cyan', linestyle='-', linewidth=0.5, which='major')

def label_bars(ax, bars, text_format):
    # Get y-axis height to calculate label position from.
    (y_bottom, y_top) = ax.get_ylim()
    y_height = y_top - y_bottom

    for bar in bars:
        height = bar.get_height()
        color = "white"
        text = text_format.format(bar.get_height())
        text_x = bar.get_x() + bar.get_width() / 2
        p_height = (height / y_height)
        
        if p_height > 0.5: 
            label_position = (y_height * 0.01)
        else:
            label_position = height + (y_height * 0.18)

        ax.text(text_x, label_position, text,  ha='center', color=color)

value_format = "{:.1f}%"
label_bars(ax, bars, value_format)
label_bars(ax, bars2, value_format)

fig.savefig("Fig2.png")
```

## Trends

* Capomulin is clearly the best of these drugs in terms of tumor volume, metastatic sites, and survival rate.
* Infubinol and Ketapril had no discernable effect on average tumor volume and shared almost identical results with the Placebo.
* Infubinol decreased the amount of metastatic sites more than Ketapril & Placebo, but not as much Capomulin.
